# Plotting Function

```python
def plotting(dbscan_full, dbscan_partial, kmeans_full, kmeans_partial, plotting_location, plotting_lim, saving_directory):
    
    legendlist=['Benign Control Cost', 'Without Appliance Triggering Attack','With Appliance Triggering Attack' ]
    markers=['v','o','s', '<']
    fig= plt.figure(figsize=(8,6.5))
    plt.plot(list(range(1, 21)), dbscan_full, marker=markers[0], linewidth=3, markersize=12)
    plt.plot(list(range(1, 21)), dbscan_partial, marker=markers[1], linewidth=3, markersize=12)
    plt.plot(list(range(1, 21)), kmeans_full, marker=markers[2], linewidth=3, markersize=12)
    plt.plot(list(range(1, 21)), kmeans_partial, marker=markers[3], linewidth=3, markersize=12)
    plt.legend(legendlist,loc='upper center',fontsize="20")
    plt.yticks(fontsize="20")
    plt.xticks(fontsize="20")
    plt.ylim(plotting_lim)
    #plt.title(plt_title, fontsize = "16")
    plt.xlabel("Number of Episode",fontsize="24")
    plt.ylabel("Expected Return",fontsize="24")
    plt.grid(True)
    plt.savefig(saving_directory, bbox_inches='tight')
```

# Reading attack scheduled

```python
plotting_location = 'upper center'
saving_directory = str(parent_directory) + '\plots\\Eval-G_House-A_Occ-1.pdf'
plotting_lim = [2000, 5500]

plotting(house_A_occ_1_dbscan_full, house_A_occ_1_dbscan_partial, house_A_occ_1_kmeans_full, house_A_occ_1_kmeans_partial, plotting_location, plotting_lim,  saving_directory)
```
