# Plotting Function

```python
def plotting(dbscan_full, dbscan_partial, kmeans_full, kmeans_partial, plotting_location, plotting_lim, saving_directory):
    
    legendlist=['ADM: DBSCAN, Knowledge: Full', 'ADM: DBSCAN, Knowledge: Partial','ADM: K-Means Clustering, Knowledge: Full', 'ADM: K-Means Clustering, Knowledge: Partial',]
    markers=['v','o','s', '<']
    fig= plt.figure(figsize=(8,6.5))
    plt.plot(list(range(1, 21)), dbscan_full, marker=markers[0], linewidth = 3, markersize = 10, color = 'lightgray', linestyle = '--')
    plt.plot(list(range(1, 21)), dbscan_partial, marker=markers[1], linewidth = 3, markersize = 10, color = 'gray', linestyle = ':')
    plt.plot(list(range(1, 21)), kmeans_full, marker=markers[2], linewidth = 3, markersize = 10, color = 'darkgray', linestyle = '-.')
    plt.plot(list(range(1, 21)), kmeans_partial, marker=markers[3], linewidth = 3, markersize = 10, color = 'black', linestyle='--', dashes=(5, 2))
    plt.legend(legendlist,loc='upper center',fontsize="16")
    plt.yticks(fontsize="20")
    plt.xticks([4, 8, 12, 16, 20],fontsize="20")
    plt.ylim(plotting_lim)
    plt.xlabel("Number of Episodes",fontsize="24")
    plt.ylabel("Expected Return",fontsize="24")
    plt.grid(True)
    plt.savefig(saving_directory, bbox_inches='tight')
```
